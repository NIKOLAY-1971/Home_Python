```python
# Даны две функции:
# f(x) = x^3 - 50x и g(x) = -x^4 + 88x^2 - 241
# Требуется:
# 1. Найти точки пересечения этих функций.
# 2. Построить графики функций в одной системе координат
# 3. Построить график общей функции
```

```python
from sympy import *
```

```python
x = Symbol('x')
fx = x**3-50*x
gx = -x**4+88*x**2-241
```

```python
# 1. Найти точки пересечения этих функций.
```

```python
ex = fx-gx
i_ex = solve(ex)
```

```python
i_x = []
for i in i_ex:
    i_x.append(i.evalf(4, chop = True))
i_x
```

```python
i_y = []
for i in i_ex:
    i_y.append(fx.subs(x,i.evalf(4, chop = True)))
i_y
```

```python
print('Точки пересечения:')
for i, (x, y) in enumerate(zip(i_x, i_y)):
    print(f'({x}, {y})')
```

```python
# 2. Построить графики функций в одной системе координат
plot(fx, gx)
```
